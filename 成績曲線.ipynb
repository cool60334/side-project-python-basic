```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# 載入成績csv檔案，取得成績

```python
data = []

with open("學生成績2.csv", encoding = "utf-8-sig") as infile:
  
  # 取得學生成績
  text = infile.read().split() # text是一個list型態
  # print(text)  # ["學生1名稱及成績", "學生2名稱及成績",...]
  captions = text[0].split(",")  # 取得各個標題
  # print(captions)


  for e in text[1:]:
    stu = e.split(",")  # 將每個學生名稱及成績用逗點分隔變成list
    # print(stu)  # ["學生1名稱", "成績1", "成績2",...]

    scores = [float(sc) for sc in stu[1:]]  # 將文字改成浮點數

    # print([stu[0]])  # ['Tom']
    data.append([stu[0]] + scores)
    
# data  # [['Tom', 100.0, 90.0, 100.0, 80.0, 85.0, 91.5],...]
```

# 計算平均分數

```python
avg = []

for i in range(1, len(data[0])):
  all_scores = [data[stu][i] for stu in range(len(data))]  # 取得所有成績
  avg.append(sum(all_scores)/len(all_scores))  # 計算平均

avg
```

# 繪製曲線圖表

```python
for stu in data:
  
  name = stu[0]  # 學生名稱
  scores = stu[1:]  # 學生成績

  plt.clf()  # 設定全新圖表
  plt.plot(scores, marker = "o", label = "Your Score")  # 繪製各別分數折線圖，(Y軸的值, 點的樣式, 折線圖名稱)
  plt.plot(avg, marker = "x", label = "Class Average")  # 繪製平均分數折線圖
  plt.legend()  # 顯示折線圖名稱

  plt.title(name)  # 設定圖表的標題

  # range(len(scores)) -> 表示X軸的刻度數量
  # captions[1:] -> 表示刻度名稱
  plt.xticks(range(len(scores)), captions[1:])


  plt.xlabel("Items")  # X軸標題
  plt.ylabel("Scores")  # Y軸標題

  plt.ylim(0, 110)  # 將Y軸的值設定在0到110之間

  plt.tight_layout()  # 把圖表限制在圖片中

  plt.savefig(name + ".png")  # 將圖表存為png檔
  plt.show()
```
